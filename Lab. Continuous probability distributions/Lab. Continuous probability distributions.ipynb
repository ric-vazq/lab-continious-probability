## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
uniform_distribution = scipy.stats.uniform()
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def gen_uniform_rdn_num_dist(bottom, ceiling, count):
    uniform_dist = scipy.stats.uniform(loc=bottom, scale=ceiling - bottom)
    return uniform_dist.rvs(size=count)
```

```python
bottom1, ceiling1, count1 = 10, 15, 100

bottom2, ceiling2, count2, = 10, 60, 1000

rdm_num1 = gen_uniform_rdn_num_dist(bottom1, ceiling1, count1)
rdm_num2 = gen_uniform_rdn_num_dist(bottom2, ceiling2, count2)
```

```python
# Plotting the histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for the first set of random numbers
ax1.hist(rdm_num1, bins=10, edgecolor='black')
ax1.set_title('Uniform Distribution (10, 15, 100)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')

# Histogram for the second set of random numbers
ax2.hist(rdm_num2, bins=10, edgecolor='black')
ax2.set_title('Uniform Distribution (10, 60, 1000)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# The one with a higher count is more uniform than the one with a lower count. 
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

```python
from scipy.stats import norm 
```

```python
def gen_normal_rdm_num(mean, std_dev, count):
    normal_dist = scipy.stats.norm(loc=mean, scale=std_dev)
    return normal_dist.rvs(size=count)
```

```python
mean1, std_dev1, count1 = 10, 1, 1000
mean2, std_dev2, count2 = 10, 50, 1000
```

```python
normal_numbers1 = gen_normal_rdm_num(mean1, std_dev1, count1)
normal_numbers2 = gen_normal_rdm_num(mean2, std_dev2, count2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for the first set of random numbers
ax1.hist(normal_numbers1, bins=30, edgecolor='black')
ax1.set_title('Normal Distribution (mean=10, std_dev=1, count=1000)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')

# Histogram for the second set of random numbers
ax2.hist(normal_numbers2, bins=30, edgecolor='black')
ax2.set_title('Normal Distribution (mean=10, std_dev=50, count=1000)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

```python
def gen_exponential_rdm_num(mean, size):
    return np.random.exponential(scale=mean, size=size)

sequence1 = gen_exponential_rdm_num(10, 10)
sequence2 = gen_exponential_rdm_num(10, 100)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.hist(sequence1, bins=100, edgecolor='black')
ax1.set_title('Exponential Distribution (mean=10, size=10)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')

ax2.hist(sequence2, bins=100, edgecolor='black')
ax2.set_title('Exponential Distribution (mean=10, size=100)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
mean_time = 10
lambda_param = 1 / mean_time

time_threshold = 15

probability = 1 - np.exp(-lambda_param * time_threshold)
probability
```

What is the probability that the customer will spend more than 15 minutes

```python
probability_more_than_15 = 1 - probability
probability_more_than_15
```
